##**데이터베이스 Database**

체계적으로 데이터 저장, 관리하는 시스템
- **관계형 데이터베이스 (RDBMS, Relational Database Management System)**
    - 데이터가 표(테이블) 형태로 저장, 테이블 간의 관계 정의하여 관리
- **비관계형 데이터베이스 (NoSQL, Not Only SQL)**
    - JSON, 키-값, 그래프 형태로 데이터 저장, 스키마가 고정되지 않아 유연한 데이터 처리 가능

##**사용자 계정 관리**

```python
# 사용자 계정 생성 및 비밀번호 설정
CREATE USER 'user1'@'localhost' IDENTIFIED BY '1234';
```

```python
# 사용자 삭제
DROP USER 'user1'@'localhost';
```

image.png

##**비밀번호 변경**

```python
# 특정 사용자의 비밀번호 변경
ALTER USER 'user1'@'localhost' IDENTIFIED BY 'newpassword123';
```

```python
# 로그인한 사용자가 직접 비밀번호 변경
SET PASSWORD = PASSWORD('newpassword123');
```



---

##**사용자 권한 관리**

```python
# 모든 권한 부여
GRANT ALL PRIVILEGES ON mydb.* TO 'user1'@'localhost';
```

```python
# 특정 권한 부여
GRANT SELECT, INSERT, UPDATE ON mydb.* TO 'user1'@'localhost';
```

```python
# 특정 권한 제거(회수)
REVOKE INSERT, UPDATE ON mydb.* FROM 'user1'@'localhost';
```

**권한 변경사항 적용 (필수 실행)**

```python
FLUSH PRIVILEGES;
```
