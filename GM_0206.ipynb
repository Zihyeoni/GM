## **클래스 (Class)**

### **Method overiding (오버라이딩)**
서브 클래스가 부모 클래스의 method 재정의

```python
class Parent:
    def display(self):
        print("부모 클래스의 display 메서드입니다.")

class Child(Parent):
    def display(self):
        print("자식 클래스의 display 메서드입니다.")

child = Child()

child.display()
```

### **클래스 method & 정적 method**
클래스 method: 클래스 자체를 인자로 받는 method

정적 method: 클래스나 인스턴스와 무관하게 동작하는 method

```python
class MyClass:
  class_variable = "클래스 변수"

  def __init__(self, value):
      self.instance_variable = value

  @classmethod
  def class_method(cls):
      return cls.class_variable

  @staticmethod
  def static_method():
      return "정적 메서드"

# 클래스 메서드 호출
print(MyClass.class_method())

# 정적 메서드 호출
print(MyClass.static_method())
```

### **데이터 캡슐화**
객체의 속성을 외부에서 직접 접근하지 못하도록 보호하는 기법

```python
class EncapsulatedClass:
    def __init__(self, value):
        self.__hidden_variable = value  # 프라이빗 변수

    def get_hidden_variable(self):
        return self.__hidden_variable

    def set_hidden_variable(self, value):
        self.__hidden_variable = value

# 인스턴스 생성
obj = EncapsulatedClass(10)

# 프라이빗 변수 접근
print(obj.get_hidden_variable())
obj.set_hidden_variable(20)
print(obj.get_hidden_variable())
```

### **특수 method**
python에서 미리 정의된 method (예: image.png)

```python
class Book:
  def __init__(self, title, author, pages):
      self.title = title
      self.author = author
      self.pages = pages

  def __str__(self):
      return f"{self.title} by {self.author}"

  def __len__(self):
      return self.pages

book = Book("The Great Gatsby", "F. Scott Fitzgerald", 180)

# 특수 메서드 호출
print(book)
print(len(book))
```

Car 클래스를 정의하고, make, model, year 속성을 초기화하는 생성자를 작성하세요. 또한, start, stop 메서드를 추가하여 자동차의 상태를 출력하는 기능을 구현하세요.

```python
class Car:
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year
    self.is_running = False

  def start(self):
    self.is_running = True
    print(f"{self.year} {self.make} {self.model}가 시작되었습니다.")

  def stop(self):
    self.is_running = False
    print(f"{self.year} {self.make} {self.model}가 멈췄습니다.")

car = Car("Toyota", "Corolla", 2024)
car.start()
car.stop()
```

Rectangle 클래스를 정의하고, width, height 속성을 초기화하는 생성자를 작성하세요. 또한, area, perimeter 메서드를 추가하여 직사각형의 넓이와 둘레를 계산하는 기능을 구현하세요.

```python
class Rectangle:
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

  def perimeter(self):
    return (self.width + self.height) * 2

rec = Rectangle(5, 8)
print(f"직사각형 넓이: {rec.area()}")
print(f"직사각형 넓이: {rec.perimeter()}")
```

BankAccount 클래스를 정의하고, owner, balance 속성을 초기화하는 생성자를 작성하세요. 또한, deposit, withdraw 메서드를 추가하여 입금과 출금 기능을 구현하세요.

```python
class BankAccount:
  def __init__(self, owner, balance=0):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    print(f"{amount}원이 입금되었습니다. 잔액: {self.balance}원")

  def withdraw(self, amount):
    if amount <= self.balance:
      self.balance -= amount
      print(f"{amount}원이 출금되었습니다. 잔액: {self.balance}원")
    else:
      print("잔액이 부족합니다.")

account = BankAccount("HyunJi", 1000)
account.deposit(500)
account.withdraw(70)
account.withdraw(2000)
```

Student 클래스를 정의하고, name, student_id, grades 속성을 초기화하는 생성자를 작성하세요. 또한, add_grade, average_grade 메서드를 추가하여 성적을 추가하고 평균 성적을 계산하는 기능을 구현하세요.

```python
class Student:
  def __init__(self, name, student_id):
    self.name = name
    self.student_id = student_id
    self.grades = []

  def add_grade(self, grade):
    self.grades.append(grade)

  def average_grade(self):
    ave = sum(self.grades) / len(self.grades)
    print(f"{self.name}[{self.student_id}]의 평균 점수: {ave:.2f}")

student = Student("HyunJi", 422)
student.add_grade(80)
student.add_grade(90)
student.add_grade(95)
student.average_grade()
```

### Todo list: todo.py & main.py 이용한 Todo list app

```python
# main.py

from todo import TodoManager

def display_menu():
    print("\n할 일 관리 프로그램")
    print("1. 할 일 추가")
    print("2. 할 일 제거")
    print("3. 할 일 목록 보기")
    print("4. 종료")

def main():
    manager = TodoManager()

    while True:
        display_menu()
        choice = input("선택: ")

        if choice == '1':
            task = input("추가할 할 일: ")
            manager.add_task(task)
        elif choice == '2':
            task = input("제거할 할 일: ")
            manager.remove_task(task)
        elif choice == '3':
            manager.show_tasks()
        elif choice == '4':
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 입력입니다. 다시 시도하세요.")

if __name__ == "__main__": #다른 모듈에서 임포트될 때 테스트 코드나 메인 함수 호출이 실행되지 않도록 하기 위함
    main()
```

```python
# todo.py

class TodoManager:
    def __init__(self):
        self.todos = []

    def add_task(self, task):
        """할 일을 목록에 추가합니다."""
        self.todos.append(task)
        print(f"'{task}'가 할 일 목록에 추가되었습니다.")

    def remove_task(self, task):
        """할 일을 목록에서 제거합니다."""
        if task in self.todos:
            self.todos.remove(task)
            print(f"'{task}'가 할 일 목록에서 제거되었습니다.")
        else:
            print(f"'{task}'는 목록에 없습니다.")

    def show_tasks(self):
        """현재 할 일 목록을 출력합니다."""
        if not self.todos:
            print("할 일 목록이 비어 있습니다.")
        else:
            print("할 일 목록:")
            for idx, task in enumerate(self.todos, start=1):
                print(f"{idx}. {task}")
```

### 로또번호 발생기: lotto_generator.py & main.py 이용한 로또 번호 발생기

```python
# main.py

from lotto_generator import LottoGenerator

def display_menu():
    print("\n로또 번호 생성기")
    print("1. 로또 번호 생성")
    print("2. 종료")

def main():
    generator = LottoGenerator()

    while True:
        display_menu()
        choice = input("선택: ")

        if choice == '1':
            numbers = generator.generate_numbers()
            print(f"생성된 로또 번호: {numbers}")
        elif choice == '2':
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 입력입니다. 다시 시도하세요.")

if __name__ == "__main__":
    main()
```

```python
# lotto_generator.py

import random

class LottoGenerator:
    def __init__(self, num_numbers=6, num_range=45):
        self.num_numbers = num_numbers
        self.num_range = num_range

    def generate_numbers(self):
        return sorted(random.sample(range(1, self.num_range + 1), self.num_numbers))
```

### **데이터 조회 및 탐색**
- **데이터 로드 및 확인**: 데이터 불러오고 기본 정보 출력
- **특정 조건 필터링**: 광고비, 매출액, 구매전환율 등의 기준으로 데이터 조회
- **기간별 조회**: 특정 날짜 범위 내 데이터 선택
- **상위 N개 데이터 조회**: 매출액 상위 10개 데이터 확인
- **광고비 대비 매출 비율 계산**: 새로운 컬럼 추가하여 광고비 대비 매출 효과 분석

```python
import pandas as pd

# 데이터 로드
file_path = "marketing_data.csv"  # 로컬 경로 사용
df = pd.read_csv(file_path, encoding="utf-8-sig")

# 날짜 데이터를 datetime 형식으로 변환
df["날짜"] = pd.to_datetime(df["날짜"], errors="coerce")

# 데이터 기본 정보 출력
print("데이터프레임 정보:")
print(df.info())

# 상위 5개 데이터 확인
print("\n데이터 미리보기:")
print(df.head())

# 광고비가 300만원 이상인 데이터 조회
high_ad_cost = df[df["광고비(만원)"] >= 300]
print("\n광고비 300만원 이상 데이터:")
print(high_ad_cost)

# 매출액이 5000만원 이상이고, 구매전환율이 5% 이상인 데이터 조회
high_sales_conversion = df[(df["매출액(만원)"] >= 5000) & (df["구매전환율(%)"] >= 5)]
print("\n매출액 5000만원 이상 & 구매전환율 5% 이상 데이터:")
print(high_sales_conversion)

# 특정 날짜 범위 조회 (예: 2024년 1월 데이터)
date_filtered = df[(df["날짜"] >= "2024-01-01") & (df["날짜"] <= "2024-01-31")]
print("\n2024년 1월 데이터:")
print(date_filtered)

# 매출액 상위 10개 데이터 조회
top_10_sales = df.nlargest(10, "매출액(만원)")
print("\n매출액 상위 10개 데이터:")
print(top_10_sales)

# 광고비 대비 매출 비율 계산 (새로운 컬럼 추가)
df["광고비 대비 매출 비율"] = df["매출액(만원)"] / df["광고비(만원)"]
print("\n광고비 대비 매출 비율 계산:")
print(df[["날짜", "광고비(만원)", "매출액(만원)", "광고비 대비 매출 비율"]].head())

# 광고비 대비 매출 비율이 가장 높은 5개 데이터 조회
top_roi = df.nlargest(5, "광고비 대비 매출 비율")
print("\n광고비 대비 매출 효과가 높은 5개 데이터:")
print(top_roi)
```

```python
# 날짜 데이터를 datetime 형식으로 변환
df["날짜"] = pd.to_datetime(df["날짜"])

# 1. 날짜별 매출액(만원) 합계 구하기
grouped_by_date = df.groupby("날짜", observed=False)["매출액(만원)"].sum()
print("날짜별 매출액 합계:\n", grouped_by_date, "\n")

# 2. 광고비(만원) 구간별 평균 방문자수 구하기
df["광고비구간"] = pd.cut(df["광고비(만원)"], bins=[0, 100, 300, 500, 1000], labels=["0-100", "100-300", "300-500", "500+"])
grouped_by_budget = df.groupby("광고비구간", observed=False)["방문자수"].mean()
print("광고비 구간별 평균 방문자수:\n", grouped_by_budget, "\n")

# 3. 구매전환율(%) 범주별 매출액 평균 구하기, pd.cut(특정 구간으로 나눌 때 사용)
df["구매전환율구간"] = pd.cut(df["구매전환율(%)"], bins=[0, 2, 5, 10], labels=["0-2%", "2-5%", "5-10%"])
grouped_by_conversion = df.groupby("구매전환율구간", observed=False)["매출액(만원)"].mean()
print("구매전환율 구간별 평균 매출액:\n", grouped_by_conversion, "\n")

# 4. SNS 공유수 그룹별 평균 광고비 구하기
df["SNS공유구간"] = pd.cut(df["SNS공유수"], bins=[0, 100, 300, 500], labels=["0-100", "100-300", "300-500"])
grouped_by_sns = df.groupby("SNS공유구간", observed=False)["광고비(만원)"].mean()
print("SNS 공유수 구간별 평균 광고비:\n", grouped_by_sns, "\n")

# 5. 날짜별 평균 구매전환율(%) 구하기
grouped_by_date_conversion = df.groupby("날짜", observed=False)["구매전환율(%)"].mean()
print("날짜별 평균 구매전환율:\n", grouped_by_date_conversion, "\n")

# 6. 광고비(만원)별 평균 방문자수 및 매출액(만원) 구하기
grouped_budget_stats = df.groupby("광고비(만원)", observed=False)[["방문자수", "매출액(만원)"]].mean()
print("광고비별 평균 방문자수 및 매출액:\n", grouped_budget_stats, "\n")

# 7. 방문자수 구간별 고객리뷰 평균 개수 구하기
df["방문자구간"] = pd.cut(df["방문자수"], bins=[0, 1000, 3000, 5000, 10000], labels=["0-1000", "1000-3000", "3000-5000", "5000+"])
grouped_by_visitors = df.groupby("방문자구간", observed=False)["고객리뷰수"].mean()
print("방문자수 구간별 평균 고객리뷰수:\n", grouped_by_visitors, "\n")

# 8. 이메일 구독자수 구간별 평균 SNS 공유수 구하기
df["이메일구독구간"] = pd.cut(df["이메일구독자수"], bins=[0, 5000, 10000, 20000], labels=["0-5000", "5000-10000", "10000-20000"])
grouped_by_subscribers = df.groupby("이메일구독구간", observed=False)["SNS공유수"].mean()
print("이메일 구독자수 구간별 평균 SNS 공유수:\n", grouped_by_subscribers, "\n")
```

####**Growth marketing 성공 분석 예시**

```python
"""
import pandas as pd

# 데이터 로드
file_path = "/content/manufacturing_growth_marketing.csv"
df = pd.read_csv(file_path, encoding="utf-8-sig")
"""
# 날짜 데이터를 datetime 형식으로 변환
gm_df["날짜"] = pd.to_datetime(gm_df["날짜"])

# 1. 광고비 대비 매출 성과 (ROAS: Return on Ad Spend)
gm_df["ROAS"] = gm_df["매출액(만원)"] / gm_df["광고비(만원)"]  # ROAS = 매출액 / 광고비
average_roas = gm_df["ROAS"].mean()

# 2. 방문자 대비 구매 전환율
average_conversion_rate = gm_df["구매전환율(%)"].mean()

# 3. B2B 문의 수 증가율 분석
gm_df["B2B 문의 변화율(%)"] = gm_df["B2B 문의 수"].pct_change() * 100
average_b2b_growth = gm_df["B2B 문의 변화율(%)"].mean()

# 4. 방문자 1인당 매출액
gm_df["1인당 매출액(만원)"] = gm_df["매출액(만원)"] / gm_df["방문자수"]
average_revenue_per_visitor = gm_df["1인당 매출액(만원)"].mean()

# 5. SNS 공유 증가율과 매출 변화율 간의 상관관계
gm_df["SNS 공유 변화율(%)"] = gm_df["SNS 공유수"].pct_change() * 100
gm_df["매출 변화율(%)"] = gm_df["매출액(만원)"].pct_change() * 100
sns_correlation = gm_df["SNS 공유 변화율(%)"].corr(gm_df["매출 변화율(%)"])

# 6. 이메일 구독자 증가율
gm_df["이메일 구독자 증가율(%)"] = gm_df["이메일 구독자수"].pct_change() * 100
average_email_growth = gm_df["이메일 구독자 증가율(%)"].mean()

# 7. B2C 판매량 증가율
gm_df["B2C 판매 변화율(%)"] = gm_df["B2C 판매 수량"].pct_change() * 100
average_b2c_growth = gm_df["B2C 판매 변화율(%)"].mean()

# 8. 평균 주문 단가 증가율
gm_df["평균 주문 단가 변화율(%)"] = gm_df["평균 주문 단가(만원)"].pct_change() * 100
average_order_price_growth = gm_df["평균 주문 단가 변화율(%)"].mean()

# 성공 여부 판단 기준
success_criteria = {
    "ROAS": average_roas >= 4,  # ROAS 4 이상이면 성공
    "구매전환율": average_conversion_rate >= 5,  # 구매전환율 5% 이상이면 성공
    "B2B 성장률": average_b2b_growth >= 2,  # B2B 문의 증가율 2% 이상이면 성공
    "방문자당 매출": average_revenue_per_visitor >= 5,  # 방문자당 매출 5만원 이상이면 성공
    "SNS-매출 상관관계": sns_correlation >= 0.3,  # SNS 공유와 매출의 상관관계가 0.3 이상이면 성공
    "이메일 구독 증가": average_email_growth >= 2,  # 이메일 구독 증가율 2% 이상이면 성공
    "B2C 판매 증가": average_b2c_growth >= 3,  # B2C 판매 증가율 3% 이상이면 성공
    "평균 주문 단가 증가": average_order_price_growth >= 1,  # 평균 주문 단가 증가율 1% 이상이면 성공
}

# 전체 성공 여부 판단, all()함수는 하나라도 false면 실패가 됨
overall_success = all(success_criteria.values())

# 결과 출력
print("성공 여부 기준 충족 여부:")
for key, value in success_criteria.items():
    print(f"{key}: {'✅ 성공' if value else '❌ 실패'}")

print("\n최종 성공 여부:", "✅ 성공" if overall_success else "❌ 실패")
```
